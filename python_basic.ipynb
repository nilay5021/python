```python
#set
f = {"apple", "banana", "orange"}
col = {"orange", "red", "green"}
f.add("mango")
print(f)
f.remove("banana")
print(f)
print(f.union(col))
print(f.intersection(col)) 
print(f.difference(col))
print(f.symmetric_difference(col))
```

```python
#loops
colors = ["red", "orange", "yellow", "green", "blue", "indigo", "violet"]
for color in colors:
    print(color)
for number in range(1, 11):
    print(number)
fruits = ["apple", "banana", "orange"]
for index, fruit in enumerate(fruits):
    print(f"At position {index}, I found a {fruit}")
```
